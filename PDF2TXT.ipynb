```python
import os
import PyPDF2

# Ganti dengan path folder Anda
folder_path = r"C:\Users\muham\OneDrive - Universitas Airlangga\Semester 6\Sistem Temu Kembali Informasi\Tugas dan Latihan\Final tugas akhir\File UU"

for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        # Membuka file PDF
        pdf_file = open(os.path.join(folder_path, filename), 'rb')
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # Membuat file teks kosong
        txt_filename = os.path.splitext(filename)[0] + '.txt'
        txt_file = open(os.path.join(folder_path, txt_filename), 'w', encoding='utf-8')

        # Mengekstrak teks dari setiap halaman PDF
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            txt_file.write(page.extract_text())

        # Menutup file PDF dan teks
        pdf_file.close()
        txt_file.close()
        print(f'Teks dari {filename} telah diekstrak ke {txt_filename}')

print('Selesai mengekstrak teks dari semua file PDF.')
```
